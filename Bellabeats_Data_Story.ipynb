```python
#Overview of the Data
import pandas as pd

# Define the path to the files
file_paths = {
    'dailyActivity': 'dailyActivity.csv',
    'dailyCalories': 'dailyCalories.csv',
    'dailyIntensities': 'dailyIntensities.csv',
    'dailySteps': 'dailySteps.csv',
    'heartrate_seconds': 'heartrate_seconds.csv',
    'sleepDay': 'sleepDay.csv',
    'weightLogInfo': 'weightLogInfo.csv'
}

# Read each file into a pandas DataFrame
dataframes = {name: pd.read_csv(path) for name, path in file_paths.items()}

# Explore each DataFrame
for name, df in dataframes.items():
    print(f"Exploring {name} dataset")
    display(df.head())  # Show the first few records
    display(df.info())  # Show a summary of the dataset
    display(df.describe())  # Present basic statistics of numerical columns
    print("\n" + "-"*50 + "\n")  # Print a separator between datasets
```

```python
# Smart Usage Data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dailyActivity data into a DataFrame
daily_activity = pd.read_csv('dailyActivity.csv')

# Basic summary statistics
summary_stats = daily_activity.describe()

# Histogram of Total Steps taken
plt.figure(figsize=(10, 5))
sns.histplot(daily_activity['StepTotal'], bins=30, kde=False)
plt.title('Distribution of Total Steps')
plt.xlabel('Total Steps')
plt.ylabel('Count')
plt.grid(True)
plt.show()

# format string and convert the 'ActivityDay' column to datetime
daily_activity['ActivityDay'] = pd.to_datetime(daily_activity['ActivityDay'], format='%m/%d/%y')
daily_activity.sort_values('ActivityDay', inplace=True)

plt.figure(figsize=(15, 7))
plt.plot(daily_activity['ActivityDay'], daily_activity['StepTotal'])
plt.title('Total Steps Over Time')
plt.xlabel('Date')
plt.ylabel('Total Steps')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Output the summary statistics
print(summary_stats)
```

```python
# Customer Feedback
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the customer reviews data
customer_reviews = pd.read_csv('customerReviews.csv')

# rating column is named 'rating' and is already in a numerical format
customer_reviews['rating_numerical'] = customer_reviews['rating'].astype(float)

# Plotting the distribution of ratings
plt.figure(figsize=(10, 5))
sns.countplot(x='rating_numerical', data=customer_reviews)
plt.title('Distribution of Review Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
# Customer Feedback
from textblob import TextBlob

# Perform sentiment analysis
customer_reviews['sentiment'] = customer_reviews['text'].apply(lambda x: TextBlob(x).sentiment.polarity)

# Visualize the sentiment distribution
plt.figure(figsize=(10, 5))
sns.histplot(customer_reviews['sentiment'], bins=30, kde=False)
plt.title('Distribution of Review Sentiments')
plt.xlabel('Sentiment Score')
plt.ylabel('Count')
plt.show()
```

```python
# Customer Behavior Analysis
# Average daily steps by day of the week
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dailyActivity data
daily_activity = pd.read_csv('dailyActivity.csv')  # Adjust path as necessary
daily_activity['ActivityDay'] = pd.to_datetime(daily_activity['ActivityDay'], format='%m/%d/%y')
daily_activity['DayOfWeek'] = daily_activity['ActivityDay'].dt.day_name()

# Calculate average steps by day of the week
avg_steps_by_day = daily_activity.groupby('DayOfWeek')['StepTotal'].mean().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

# Plotting average steps by day of the week
plt.figure(figsize=(10, 6))
avg_steps_by_day.plot(kind='bar', color='skyblue')
plt.title('Average Daily Steps by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Steps')
plt.xticks(rotation=45)
plt.show()
```

```python
# Customer Behavior Analysis
# Average calories burned by day of the week
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dailyCalories data
daily_calories = pd.read_csv('dailyCalories.csv')  # Adjust path as necessary
daily_calories['ActivityDay'] = pd.to_datetime(daily_calories['ActivityDay'], format='%m/%d/%Y')
daily_calories['DayOfWeek'] = daily_calories['ActivityDay'].dt.day_name()

# Calculate average calories burned by day of the week
avg_calories_by_day = daily_calories.groupby('DayOfWeek')['Calories'].mean().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

# Plotting average calories burned by day of the week
plt.figure(figsize=(10, 6))
avg_calories_by_day.plot(kind='bar', color='orange')
plt.title('Average Daily Calories Burned by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Calories Burned')
plt.xticks(rotation=45)
plt.show()
```

```python
# Customer Behavior Analysis
# Analyzing sleep patterns
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the sleepDay data
sleep_day = pd.read_csv('sleepDay.csv')  # Adjust path as necessary

# Since the date format includes time, used '%m/%d/%Y' and ignore the time part
sleep_day['SleepDay'] = pd.to_datetime(sleep_day['SleepDay']).dt.date

# Convert back to datetime without the time component for easier handling
sleep_day['SleepDay'] = pd.to_datetime(sleep_day['SleepDay'], format='%Y-%m-%d')

sleep_day['DayOfWeek'] = sleep_day['SleepDay'].dt.day_name()

# 'TotalMinutesAsleep' is the column I want to analyze
avg_sleep_by_day = sleep_day.groupby('DayOfWeek')['TotalMinutesAsleep'].mean().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

plt.figure(figsize=(10, 6))
avg_sleep_by_day.plot(kind='bar', color='purple')
plt.title('Average Sleep Minutes by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Minutes Asleep')
plt.xticks(rotation=45)
plt.show()
```

```python
# Trend Analysis
# Trend analysis overtime

daily_activity['ActivityDay'] = pd.to_datetime(daily_activity['ActivityDay'], format='%m/%d/%y')

# Grouping by date and calculating average steps
daily_avg_steps = daily_activity.groupby('ActivityDay')['StepTotal'].mean()

plt.figure(figsize=(15, 7))
daily_avg_steps.plot()
plt.title('Daily Average Steps Over Time')
plt.xlabel('Date')
plt.ylabel('Average Steps')
plt.show()
```

```python
# Trend Analysis
# Monthly trend analysis

# Grouping by month and calculating average steps
monthly_avg_steps = daily_activity.resample('M', on='ActivityDay')['StepTotal'].mean()

plt.figure(figsize=(15, 7))
monthly_avg_steps.plot(marker='o', linestyle='-')
plt.title('Monthly Average Steps Over Time')
plt.xlabel('Month')
plt.ylabel('Average Steps')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Trend Analysis
# Analyzing sesonal trends or response on specific events

# Highlighting New Year Period
new_year_start = pd.to_datetime('01/01/2021')  
new_year_end = pd.to_datetime('01/31/2021')  

plt.figure(figsize=(15, 7))
monthly_avg_steps.plot(marker='o', linestyle='-', label='Monthly Avg Steps')
plt.axvline(x=new_year_start, color='r', linestyle='--', lw=2, label='New Year Start')
plt.axvline(x=new_year_end, color='g', linestyle='--', lw=2, label='End of January')
plt.title('Monthly Average Steps Over Time with New Year Highlighted')
plt.xlabel('Month')
plt.ylabel('Average Steps')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Health & wellness insights
# Activity heatmap
# daily_activity is loaded and ActivityDate is converted to datetime
daily_activity['DayOfWeek'] = daily_activity['ActivityDay'].dt.day_name()
daily_activity['Week'] = daily_activity['ActivityDay'].dt.isocalendar().week

# Pivot table for heatmap
activity_heatmap_data = daily_activity.pivot_table(index='DayOfWeek', columns='Week', values='StepTotal', aggfunc='mean')

plt.figure(figsize=(15, 7))
sns.heatmap(activity_heatmap_data, cmap='viridis')
plt.title('Average Daily Steps Heatmap by Week')
plt.xlabel('Week of the Year')
plt.ylabel('Day of the Week')
plt.show()
```
